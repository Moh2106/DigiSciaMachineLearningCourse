## Standardisation et normalisation

**Standardisation** et **normalisation** sont deux techniques de prétraitement des données couramment utilisées en machine learning pour ajuster les valeurs numériques des caractéristiques (features) afin qu'elles soient sur une échelle similaire, facilitant ainsi l'entraînement des modèles. Ces techniques sont particulièrement utiles lorsque les algorithmes de machine learning sont sensibles à l'échelle des données.

### 1. **Standardisation**

La **standardisation** consiste à transformer les données de manière à ce qu'elles aient une moyenne (mean) de 0 et un écart-type (standard deviation) de 1. La formule de standardisation pour une caractéristique \( x \) est :

\[
x_{standard} = \frac{x - \mu}{\sigma}
\]

où :
- \( \mu \) est la moyenne de la caractéristique \( x \).
- \( \sigma \) est l'écart-type de la caractéristique \( x \).

#### **Utilité :**
- **Algorithmes linéaires** : Les algorithmes comme la régression linéaire, la régression logistique, et les machines à vecteurs de support (SVM) fonctionnent mieux lorsque les données sont standardisées, car ces modèles supposent souvent que les données sont centrées autour de zéro avec une variance similaire.
- **Distance et métriques** : Dans les méthodes basées sur les distances, comme les K-plus proches voisins (K-NN), la standardisation est cruciale car les différences d'échelle peuvent autrement dominer les calculs de distance.

### 2. **Normalisation**

La **normalisation** consiste à redimensionner les données pour qu'elles se situent dans une plage spécifique, généralement [0, 1]. La formule de normalisation pour une caractéristique \( x \) est :

\[
x_{normalized} = \frac{x - x_{min}}{x_{max} - x_{min}}
\]

où :
- \( x_{min} \) est la valeur minimale de la caractéristique \( x \).
- \( x_{max} \) est la valeur maximale de la caractéristique \( x \).

#### **Utilité :**
- **Algorithmes basés sur la distance** : Les algorithmes comme K-NN ou les réseaux de neurones bénéficient souvent de la normalisation, car elle permet d'éviter que certaines caractéristiques dominent d'autres uniquement en raison de leur échelle.
- **Réseaux de neurones** : Dans les réseaux de neurones, la normalisation peut accélérer la convergence lors de l'entraînement en s'assurant que les gradients restent dans une plage cohérente.

### **Résumé de l'utilité :**

- **Meilleure performance des modèles** : En standardisant ou normalisant les données, on améliore la performance des modèles de machine learning, surtout ceux qui sont sensibles à l'échelle des caractéristiques.
- **Convergence plus rapide** : Cela permet également à certains algorithmes d'apprendre plus rapidement et plus efficacement.
- **Équilibrage des caractéristiques** : Ces techniques équilibrent l'impact des différentes caractéristiques, évitant que certaines dominent les autres à cause de leurs différentes échelles.

En résumé, la standardisation et la normalisation sont des étapes cruciales dans le prétraitement des données, particulièrement lorsque les caractéristiques des données ont des unités ou des échelles très différentes.

### Exemple pratique de la standarisation et de la normalisation avec sklearn
- La standardisation avec StandardScaler transforme vos données pour qu'elles aient une moyenne de 0 et un écart-type de 1

```python
# importation des différentes bibliothèques
import numpy as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# importer le dataset d'Iris
df = pd.read_csv("datasets/iris.csv")
```

```python
# Vue sur les 5 premiers éléments de la dataset
df.head()
```

```python
# Extraction des features (Caractéristiques)
features = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
features
```

```python
sns.pairplot(features)
plt.show()
```

```python
### La standarisation
# Initialiser le scaler
scaler = StandardScaler()
```

```python
features_standard = scaler.fit_transform(features)
```

```python
features_standard
```

```python
features_col = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df_standard = pd.DataFrame(features_standard, columns=features_col)
```

```python
df_standard 
```

```python
df_standard.describe()
```

```python
sns.pairplot(df_standard)
plt.show()
```

### Normalisation example pratique

```python
# Initialisation de la fonction de normalisation
normalized = MinMaxScaler()
```

```python
features_normalized = normalized.fit_transform(features)
features_normalized
```

```python
df_normalized = pd.DataFrame(features_normalized, columns=features_col)
```

```python
sns.pairplot(df_normalized)
```
