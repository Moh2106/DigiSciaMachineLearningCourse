### Cours détaillé sur l'algorithme K-Means

#### Introduction à K-Means

L'algorithme **K-Means** est une technique d'**apprentissage non supervisé** qui permet de regrouper des données en différentes **clusters** (ou groupes) sur la base de caractéristiques similaires. Il est principalement utilisé pour résoudre des problèmes de **clustering**, où les données ne possèdent pas de labels prédéfinis. L'objectif principal de K-Means est de diviser un ensemble de données en K groupes distincts, en minimisant la variance au sein de chaque groupe.

#### Principe de base

L'algorithme K-Means repose sur un processus itératif qui permet de diviser un ensemble de données en **K clusters**. Chaque cluster est représenté par son **centre de gravité** (appelé **centroïde**), et l'algorithme vise à minimiser la somme des distances entre chaque point de données et le centroïde de son cluster.

#### Étapes de l'algorithme

Voici les étapes principales de l'algorithme K-Means :

1. **Initialisation** :
   - Choisir aléatoirement K points dans l'ensemble de données pour servir de **centroïdes initiaux**. Ces points représentent les centres des clusters.
   
2. **Assignation** :
   - Pour chaque point de données, calculer la distance entre ce point et chaque centroïde. Le point est ensuite assigné au cluster dont le centroïde est le plus proche.
   
3. **Recalcul des centroïdes** :
   - Après l'étape d'assignation, recalculer la position du centroïde de chaque cluster en prenant la **moyenne** des coordonnées des points assignés à ce cluster.
   
4. **Itération** :
   - Répéter les étapes d'assignation et de recalcul des centroïdes jusqu'à ce que la position des centroïdes **cesse de changer** ou que la variation soit négligeable. À ce moment, l'algorithme est **convergé**.
   
5. **Résultat final** :
   - Chaque point de données est associé à un cluster, et chaque cluster possède un centroïde qui représente le centre de gravité de ses points.

#### Illustration de l'algorithme K-Means

Imaginons un exemple simple avec des données en 2D. L'algorithme fonctionnerait comme suit :

1. **Initialisation** :
   - Supposons que nous avons 3 clusters à créer (K = 3).
   - L'algorithme sélectionne aléatoirement 3 points comme centroïdes.

2. **Assignation** :
   - Chaque point de l'ensemble de données est assigné à l'un des trois clusters en fonction de la proximité avec les centroïdes.

3. **Recalcul des centroïdes** :
   - Les centroïdes sont déplacés pour être au centre des points qui leur sont assignés.

4. **Itération** :
   - Les étapes d'assignation et de recalcul sont répétées jusqu'à convergence.

#### Mesure de distance

L'algorithme K-Means utilise généralement la **distance euclidienne** pour mesurer la proximité entre les points de données et les centroïdes. La distance euclidienne entre deux points \( A(x_1, y_1) \) et \( B(x_2, y_2) \) est donnée par :

\[
d(A, B) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
\]

D'autres mesures de distance comme la **distance de Manhattan** ou la **distance cosinus** peuvent également être utilisées en fonction des besoins.

#### Choix de K

Le choix du nombre de clusters K est crucial pour obtenir des résultats pertinents. Si K est trop petit, les clusters peuvent regrouper des points qui sont trop dissemblables ; si K est trop grand, l'algorithme peut créer des clusters redondants. Le **méthode du coude** (elbow method) est une technique populaire pour déterminer la valeur optimale de K.

La méthode du coude consiste à tracer un graphique de la **somme des carrés des erreurs** (SSE) en fonction du nombre de clusters. Le SSE diminue avec l'augmentation de K, mais à un certain point, l'amélioration devient marginale, formant un "coude". Ce point représente souvent une bonne estimation de K.

#### Avantages et inconvénients de K-Means

**Avantages** :
- Facile à implémenter et à comprendre.
- Convergence rapide, surtout pour des petites dimensions.
- Bien adapté aux jeux de données de grande taille.

**Inconvénients** :
- Nécessite de définir K à l'avance.
- Sensible aux valeurs aberrantes et au bruit.
- Peut converger vers des solutions locales (dépend de l'initialisation des centroïdes).
- Fonctionne mieux avec des clusters de forme sphérique.

#### Variantes de K-Means

Il existe plusieurs variantes de l'algorithme K-Means pour pallier ses limitations, comme :

- **K-Means++** : Utilise une méthode intelligente pour initialiser les centroïdes afin d'améliorer la qualité des clusters finaux.
- **Mini-Batch K-Means** : Utilise des sous-échantillons pour améliorer la vitesse sur de grands ensembles de données.

#### Implémentation en Python

Voici un exemple simple d'implémentation de K-Means en utilisant la bibliothèque `scikit-learn` :

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Création d'un jeu de données
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Application de l'algorithme K-Means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Affichage des clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.show()
```

Dans cet exemple :
- Nous générons un ensemble de données simulé avec 4 clusters.
- Nous appliquons l'algorithme K-Means pour détecter ces clusters.
- Enfin, nous visualisons les points de données et les centroïdes avec `matplotlib`.

#### Conclusion

L'algorithme K-Means est un outil puissant pour le clustering, mais son efficacité dépend beaucoup du choix de K et de l'initialisation des centroïdes. Il est largement utilisé dans de nombreux domaines tels que la segmentation d'images, le marketing (segmentation de la clientèle), et la réduction de dimensions pour des données de haute dimension.

Il est recommandé d'explorer ses variantes comme **K-Means++** ou d'utiliser des techniques avancées comme **DBSCAN** pour des jeux de données plus complexes.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Création d'un jeu de données
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Application de l'algorithme K-Means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Affichage des clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.show()
````
